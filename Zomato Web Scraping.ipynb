```python
from bs4 import BeautifulSoup
import pandas as pd
import sys
import time
from bs4 import BeautifulSoup
import requests
```

```python
# importing the data from website

with open(r"O:\Youtube Lectures Practise\Web Scraping\Nightlife Restaurants in Baner - Zomato_1.html", encoding = 'utf') as html:
    content = html.read()
```

```python
content
```

```python
soup = BeautifulSoup(content, 'lxml')
```

```python
soup
```

```python
pub_name = soup.find_all('h4', class_='sc-1hp8d8a-0 sc-lffWgi flnmvC')
```

```python
pub_name
```

```python
# Storing the names of restaurants in a list

a = []
for i in pub_name:
    names = i.text.strip()
    a.append(names)
a
```

```python
len(pub_name)
```

```python
ratings = soup.find_all('div', class_='sc-1q7bklc-1 cILgox')
```

```python
ratings
```

```python
len(ratings)
```

```python
# Storing the ratings

b = []
for i in ratings:
    rating = i.text.strip()
    b.append(rating)
b
```

```python
cuisine_types = soup.find_all('p', class_='sc-1hez2tp-0 sc-fGSyRc eWQqcH')
```

```python
cuisine_types
```

```python
len(cuisine_types)
```

```python
# Storing the cuisines

c = []
for i in cuisine_types:
    cuisine_type = i.text.strip()
    c.append(cuisine_type)
c
```

```python
pricing = soup.find_all('p', class_='sc-1hez2tp-0 sc-fGSyRc dMkcNo')
```

```python
pricing
```

```python
len(pricing)
```

```python
# Storing the pricing 

d = []
for i in pricing:
    pricings = i.text.strip()
    d.append(pricings)
d
```

```python
location = soup.find_all('p', class_='sc-1hez2tp-0 sc-dCVVYJ tzIUc')
```

```python
location
```

```python
# Storing the location

e = []
for i in location:
    locations = i.text.strip()
    e.append(locations)
e
```

```python
# Creating a dataframe for following columns

df_zomato = pd.DataFrame()
```

```python
df_zomato['pub_names'] = a
```

```python
df_zomato['ratings'] = b
```

```python
df_zomato['cuisine_types'] = c
```

```python
df_zomato['pricing'] = d
```

```python
df_zomato['location'] = e
```
